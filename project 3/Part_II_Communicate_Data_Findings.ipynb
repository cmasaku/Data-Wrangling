## Investigation Overview


> In this investigation, I was focussing on finding out if a borrower's characteristics affects their Loan Status. The main focus was on these features: CreditGrade, EmploymentStatus, BorrowerAPR,CurrentDelinquencies, IncomeRange, IncomeVerifiable, LoanOriginalAmount, Term, IsBorrowerHomeOwner, MonthlyLoanPayment.



## Dataset Overview

> The dataset consisted of 113,937 loans with 81 variables on each loan, including loan amount, borrower rate (or interest rate), current loan status, borrower income, and many others. 61 of the variables are numeric in nature.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df=pd.read_csv('prosperLoanData.csv')
```

### Distribution of Loan Status
> Current loan status has the highest number of loans followed by completed loans. Defaulted loans comes fourth with 5018 out of 113,937 loans.

```python
base_color=sb.color_palette()[0]
type_order=df['LoanStatus'].value_counts().index
sb.countplot(data=df, y=df['LoanStatus'],color=base_color,order=type_order)
plt.title('Distribution of Loan Status')
plt.xticks(rotation=90);
```

## Relationship between IsBorrowerHomeowner vs LoanStatus

> There is an even distribution of the Loan Status and whether a borrower is a homeowner or not.

```python
sb.countplot(data=df,x='LoanStatus',hue='IsBorrowerHomeowner')
plt.title('Is Borrower HomeOwner VS Loan Status')
plt.xticks(rotation=75);
```

## Relationship between IncomeVerifiable, LoanStatus and LoanOriginalAmount

Persons whose Income is verifiable get bigger amounts of loans-Possibly because they are expected to be settle the loans with ease.

```python
g = sb.FacetGrid(data = df, col = 'IncomeVerifiable', size = 5,margin_titles=True)
g.map(sb.boxplot, 'LoanStatus', 'LoanOriginalAmount')
plt.xticks(rotation=75);
```
